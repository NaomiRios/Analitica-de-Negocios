#**Caso de Estudio**
En este documento se desarrolla y analiza un modelo Naive Bayes para la PreAprobación de créditos en una entidad financiera utilizando las variables socioeconómicos de un solicitante de crédito. De acuerdo con lo anterior las variables socioeconómicas son:

**Edad**: Es la edad que posee el solicitante al momento de solicitar un crédito.

**Ingresos**: Ingresos recibidos recibidos de una persona como resultado de una actividad comercial.

**Egresos**: Son los gastos mensuales del solicitante. Gastos de persona a cargo, hijos y servicios públicos.

**Monto(EAD)**: Indica el dinero que una persona solicita a una entidad financiera como crédito.

0. Se procede con la carga de la librería de trabajo.

```python
import numpy as np #Librería numérica por excelencia.
import pandas as pd #Librería para la comunicación con archivos excel.

#Librerías específicas.
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos del trabajo de la base de datos BaseAAA

```python
import pandas as pd # Add this line to ensure pandas is imported

nxl="/content/1. Base de Datos - SolicitantesCrédito(USD).xlsx"
XDB=pd.read_excel(nxl,sheet_name=1)
XDB.head(100) #Es para mostrar la base de datos cargada.

#Seleccionamos las variables de trabajo.
XD=XDB[ [ 'Edad','Ingresos','Egresos','Monto (EAD)']] #Variables de Trabajo
XD.head(10)
yd=XDB[['PreApr'] ] #Esta es la variable de decisión.
yd.head()
```

2. Implementamos Modelo Naives Bayes

```python
mnb=GaussianNB()
mnb.fit(XD,yd)  #Ajustar Variables Entrada - Salida.

#Mostramos las medias de las variables.

u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma)
print("'Edad','Ingresos','Egresos','Monto (EAD)'")
print(u)
print("Las desviaciones son")
print(sigma)
```

**Análisis de Resultados**

De los resultados obtenidos por el modelo podemos observar, que l acategoría de PreAprobados posee ingresos superiores a las categorías de PreNegados, esto inidca claramente la predominancia de la variable, para aprobar créditos.
Es importante mencionar que la edad es mayor para los PreAprobados que para los PreNegados, lo que indica que el banco busca personas con mayor experiencia.
Podemos decir igualmente, que los PreAprobados tienen egresos superiores a los de la categoría PreNegados, sin embargo, los PreAprobados están solicitando montos superiores  los de la categoría de PreNegados, esto es un indicador de la capacidad de este grupo para hacer frente a un crédito.

3. Se procede con la evaluación del modelo, Para la evaluación de este tipo de modelo se utiliza la matriz de confusión.
